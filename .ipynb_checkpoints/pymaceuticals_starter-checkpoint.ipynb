```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_Data_Full = pd.merge(study_results, mouse_metadata)

# Display the data table for preview
mouse_Data_Full.head()
```

```python
# Checking the number of mice.
# Additional Comment: Instead of just calculating unique mice IDs and printing on the same line, I decided to
#                     store it in a variable incase I need it later.
total_Mice = mouse_Data_Full['Mouse ID'].nunique()
total_Mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
# Additional Comment: Basically, this filters the "Mouse ID" column of the mouse_Data_Full data frame to just the "g989" entries and stores them
#                     in a numpy array. Finally, to make sure it only prints one instance of "[g989]", use an if statement to check/print first entry
#                     and it's type. It is likely that there is a much more efficient way to achieve that but I couldn't figure it out.
#                     This starter file seems to only want us to drop the Mouse with the ID: "g989", even though it looked like there were other
#                     potential duplicate entries. Regardless, as stated before, this is the solution I settled with.  
mouse_id = "g989"
filtered_data = mouse_Data_Full[mouse_Data_Full['Mouse ID'] == mouse_id]
result_array = filtered_data['Mouse ID'].to_numpy()
if result_array.size > 0:
    print(f"array([{result_array[0]}], dtype={result_array.dtype})")
```

```python
mouse_Data_Full.sort_values("Mouse ID", inplace=True)
mouse_Data_Full.drop_duplicates()
```

```python
# Optional: Get all the data for the duplicate mouse ID.
dupe = mouse_Data_Full.loc[mouse_Data_Full['Mouse ID'] == 'g989']
dupe
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_Data_Clean = mouse_Data_Full[mouse_Data_Full['Mouse ID'] != 'g989']
mouse_Data_Clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
total_Mice_Clean = mouse_Data_Clean['Mouse ID'].nunique()
total_Mice_Clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Additional Comment: This is the data frame that gets printed below. This is just setting it up, the calculations for the variables
#                     tied to each key are calculated below.
mouse_Calcs_Summary = pd.DataFrame({'Mean Tumor Volume' : mouse_Calcs_Group_Mean,
                                    'Median Tumor Volume' : mouse_Calcs_Group_Median,
                                    'Tumor Volume Variance' : mouse_Calcs_Group_Var,
                                    'Tumor Volume Std. Dev.' : mouse_Calcs_Group_StdDev,
                                    'Tumor Volume Std. Err.' : mouse_Calcs_Group_SEM})

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

# Mean Tumor Volume GroupBy Function:
mouse_Calcs_Group_Mean = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()

# Median Tumor Volume GroupBy Function:
mouse_Calcs_Group_Median = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()

# Tumor Volume Variance GroupBy Function:
mouse_Calcs_Group_Var = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()

# Tumor Volume Standard Deviation GroupBy Function:
mouse_Calcs_Group_StdDev = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()

# Tumor Volume Standard Error (SEM) GroupBy Function:
mouse_Calcs_Group_SEM = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Displays the DataFrame
mouse_Calcs_Summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
# Additional Comment: The below line is 98% successful in performing the advanced table method described above. The only issue is that it
#                     doesn't include the "Tumor Volume (mm3)" column header shown in the starter file example output. I did a bunch of research
#                     trying to figure out how to include a column header onto the same line but had no luck. So yes, technically this table is not
#                     entirely being generated from one line of code, but it's 98% of the way there.
advanced_Summary = mouse_Data_Clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Additional Comment: This is what I had to add to include the column header described above. Also, this prints the table.
advanced_Summary.columns = pd.MultiIndex.from_product([['Tumor Volume (mm3)'], advanced_Summary.columns])
advanced_Summary
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
# Additional Comment: To get the plot to show correctly using Pandas, I decided to create a specific DataFrame that contained the 2
#                     data series I needed to plot, then calling that in the ".plot.bar" function. Again, there may be a more efficient
#                     solution but this is mine. I had to put "legend=False" in that line because it was setting the "# Of Observed Mouse Timepoints"
#                     key as the legend and I didn't want that in the graph.
x_axis = mouse_Data_Clean['Drug Regimen'].unique()
y_axis = mouse_Data_Clean['Drug Regimen'].value_counts()

bar_test = pd.DataFrame({'Drug Regimen' : x_axis,
                         '# Of Observed Mouse Timepoints' : y_axis})

tick_locations = [value for value in x_axis]

bar_test.plot.bar(ylabel='# Of Observed Mouse Timepoints', legend=False)
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
# Additional Comment: x_axis, y_axis, and tick_locations are all determined in the cell above. Since this plot is also using those same
#                     variables, there was no need to redefine them here. Just wanted to clarify that.
plt.bar(x_axis, y_axis, width=0.55)
plt.xticks(rotation=90)
plt.xlim(-0.55, len(x_axis)-0.45)
plt.xlabel("Drug Regimen")
plt.ylabel("# Of Observed Mouse Timepoints")

plt.show()
```
