```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_Data_Full = pd.merge(study_results, mouse_metadata)

# Display the data table for preview
mouse_Data_Full.head()
```

```python
# Checking the number of mice.
# Additional Comment: Instead of just calculating unique mice IDs and printing on the same line, I decided to
#                     store it in a variable incase I need it later.
total_Mice = mouse_Data_Full['Mouse ID'].nunique()
total_Mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Additional Comment: This is the data frame that gets printed below. This is just setting it up, the calculations for the variables
#                     tied to each key are calculated below.
mouse_Calcs_Summary = pd.DataFrame({'Mean Tumor Volume' : mouse_Calcs_Group_Mean,
                                    'Median Tumor Volume' : mouse_Calcs_Group_Median,
                                    'Tumor Volume Variance' : mouse_Calcs_Group_Var,
                                    'Tumor Volume Std. Dev.' : mouse_Calcs_Group_StdDev,
                                    'Tumor Volume Std. Err.' : mouse_Calcs_Group_SEM})

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

# Mean Tumor Volume GroupBy Function:
mouse_Calcs_Group_Mean = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()

# Median Tumor Volume GroupBy Function:
mouse_Calcs_Group_Median = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()

# Tumor Volume Variance GroupBy Function:
mouse_Calcs_Group_Var = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()

# Tumor Volume Standard Deviation GroupBy Function:
mouse_Calcs_Group_StdDev = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()

# Tumor Volume Standard Error (SEM) GroupBy Function:
mouse_Calcs_Group_SEM = mouse_Data_Clean.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Displays the DataFrame
mouse_Calcs_Summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
# Additional Comment: The below line is 98% successful in performing the advanced table method described above. The only issue is that it
#                     doesn't include the "Tumor Volume (mm3)" column header shown in the starter file example output. I did a bunch of research
#                     trying to figure out how to include a column header onto the same line but had no luck. So yes, technically this table is not
#                     entirely being generated from one line of code, but it's 98% of the way there.
advanced_Summary = mouse_Data_Clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Additional Comment: This is what I had to add to include the column header described above. Also, this prints the table.
advanced_Summary.columns = pd.MultiIndex.from_product([['Tumor Volume (mm3)'], advanced_Summary.columns])
advanced_Summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
mouse_Timepoint_Group = mouse_Data_Clean.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
# Additional Comment: This merge results in the DataFrame having two different columns containing Timepoint data. Timepoint_x contains
#                     the max Timepoint values each Mouse ID receives, and Timepoint_y shows the progression of the Timepoint value over time.
#                     I couldn't figure out a prettier solution for what this challenge was asking for, but this worked for me.
mouse_Clean_Update = pd.merge(mouse_Timepoint_Group, mouse_Data_Clean, on='Mouse ID')
mouse_Clean_Update.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_List = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_Vol_Data = []

# Iterates through treatment_List, Store Volume data, and Calculate the IQR to quantitatively determine if there are any potential outliers.
for treatment in treatment_List:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    temp_Data = mouse_Clean_Update.loc[mouse_Clean_Update['Drug Regimen'] == treatment, :]

    # Store Volume Data
    tumor_Vol_Data.append(temp_Data['Tumor Volume (mm3)'])
    
    # Determine outliers using upper and lower bounds
    temp_LQuart = temp_Data['Tumor Volume (mm3)'].quantile(0.25)
    temp_UQuart = temp_Data['Tumor Volume (mm3)'].quantile(0.75)
    temp_IQR = temp_UQuart - temp_LQuart

    temp_Lower_Bound = temp_LQuart - (1.5*temp_IQR)
    temp_Upper_Bound = temp_UQuart + (1.5*temp_IQR)

    # Print outliers analysis for each treatment in treatment_List
    print(f"{treatment}'s potential outliers could be above {temp_Upper_Bound} and below {temp_Lower_Bound}")

# Additional Comment: This cell's example output had missing values and looked incomplete, so I don't have a good guage over what this output
#                     "should" look like or whether the results are completely correct. After double checking the calcs and tumor_Vol_Data list,
#                     the outputs seem to be correct. I opted to just show the lower and upper bounds for each Drug Regimen and say the outliers
#                     could be lower or higher than them.
```

```python
flierprops = dict(markerfacecolor='r', markersize='9')

plt.boxplot(tumor_Vol_Data, labels=treatment_List, flierprops=flierprops)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
# Additional Comment: I added a separate line to configurate the settings for outlier point visuals. I saw I could technically
#                     include that in the "plt.boxplot" main line but I separated it for ease of viewing. See my comments about the pie charts
#                     above for my discussion about potential dataset differences compared to the example plots contained in this starter file,
#                     as it seems to also apply here.
flierprops = dict(markerfacecolor='r', markersize='9')

plt.boxplot(tumor_Vol_Data, labels=treatment_List, flierprops=flierprops)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Gather Capomilin specific mouse data
cap_Data = mouse_Clean_Update[mouse_Clean_Update['Drug Regimen'] == 'Capomulin']

# Group retrieved Mouse ID's by Tumor Volume (mm3), and calculate the average tumor volume
mouse_Avg_Vol = cap_Data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merge both into a dataframe for use in plot
weight_Vol_Merge = pd.merge(mouse_Avg_Vol, cap_Data[['Mouse ID', 'Weight (g)']], on='Mouse ID')

# Use specific columns in finished dataframe to generate the scatter plot, and add some labels
plt.scatter(weight_Vol_Merge['Weight (g)'], weight_Vol_Merge['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
mouse_Corr = weight_Vol_Merge['Weight (g)'].corr(weight_Vol_Merge['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {mouse_Corr}")

regress_XVals = weight_Vol_Merge['Weight (g)']
regress_YVals = weight_Vol_Merge['Tumor Volume (mm3)']

reg_Result = linregress(regress_XVals, regress_YVals)

reg_Line = reg_Result.intercept + reg_Result.slope * regress_XVals
plt.scatter(weight_Vol_Merge['Weight (g)'], weight_Vol_Merge['Tumor Volume (mm3)'])
plt.plot(regress_XVals, reg_Line, color='red')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
